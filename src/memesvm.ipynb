```python
#imports needed for SVM model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression as LR
```

```python
#Read dataset and show first ten elements
memedata = pd.read_csv('''C:\\Users\\stefa\\Downloads\\memegenerator-dataset\\memegenerator-dataset\\memegenerator.csv''', sep='\t')
memedata.head(10)
```

```python
#Let X conatin the attributes while y contains the label
memedata_X = memedata.drop('Base Meme Name', axis=1)
memedata_y = memedata['Base Meme Name']
memedata_X.head(10)
```

```python
memedata_y.head(10)
```

```python
#Use one hot encoding to map the words into numbers for training our SVM
memeencoder = preprocessing.OneHotEncoder()
meme_y = [[y] for y in memedata_y.head(10).values]
memeencoder.fit(meme_y)
encodedmemes_X = memeencoder.transform(meme_y).toarray()
meme_X = [[x] for x in memedata_X['Alternate Text'].head(10).values]
memeencoder.fit(meme_X)
encodedmemes_y = memeencoder.transform(meme_X).toarray()
```

```python
#Create testing and training sets for our encoded dataset
X_train, X_test, y_train, y_test = train_test_split(encodedmemes_X, encodedmemes_y, test_size = 0.20)
```
